```python
train_output=pd.read_csv('ml_case_training_output.csv')
df=df_merged=df_m.merge(train_output, how='left', on='id')
```

```python
df.describe()
```

```python
# Train, Validation and Test set for Models
###############################################################################################


train_val=df.drop(columns='churn')


test=df['churn']

# print the size of each
print("train_val",train_val.shape)
print("Test",test.shape)

```

```python
#  Normalisation of the numeric columns
###############################################################################################
from sklearn.preprocessing import StandardScaler

numerical_cols = list(set(train_val.select_dtypes('number').columns) - set(['price_date','activity_new','channel_sales',
                                                                            'date_activ','date_end','date_first_activ',
                                                                            'date_modif_prod','date_renewal','has_gas',
                                                                            'origin_up']) )

print("numerical_cols",numerical_cols)


```

```python
for n in numerical_cols:
  scaler = StandardScaler()
  scaler.fit(train_val[n].values.reshape(-1,1))

  train_val_std = scaler.transform(train_val[n].values.reshape(-1,1))

  # Add it to the dataframe
  train_val_std_df = pd.DataFrame(train_val_std,index = train_val.index,columns = [n + "_std"])
  train_val = pd.concat([train_val, train_val_std_df], axis=1)


# remove the original column from the dataframe
train_val.drop(numerical_cols, inplace=True, axis=1) 

display(train_val.head())
display(test.head())
```

```python
# One hot encoding of the categorical columns on datasets
###############################################################################################
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

onehotencoder = OneHotEncoder()
# Categorical values for encoding
categorical_cols = ['price_date','activity_new','channel_sales','date_activ','date_end','date_first_activ',
                    'date_modif_prod','date_renewal','has_gas','origin_up']

for c in categorical_cols:
  onehotencoder.fit(train_val[c].values.reshape(-1,1))

  train_val_categorical = onehotencoder.transform(train_val[c].values.reshape(-1,1)).toarray()

  # Add to the dataframe back
  train_val_onehot_df = pd.DataFrame(train_val_categorical,index = train_val.index,columns = [c + str(int(i)) for i in range(train_val_categorical.shape[1])])
  train_val = pd.concat([train_val, train_val_onehot_df], axis=1)

  #test_onehot_df = pd.DataFrame(test_categorical,index = test.index,columns = [c + str(int(i)) for i in range(test_categorical.shape[1])])
  #test = pd.concat([test, test_onehot_df], axis=1)

display(train_val.head())
display(test.head())

# ### need to handle multi-collinearity seperately
```

```python
#  Create train and validation set and label set - remove not needed columns
###############################################################################################

# take only the features needed for training/testing
train_val  = train_val.iloc[:,11:]


# # print the size of each

#
from sklearn.model_selection import train_test_split

train_x,test_x,train_y,test_y=train_test_split(train_val,test,test_size=0.20,random_state=2)
print("Train",train_x.shape)
print("Test",test_x.shape)

display(train_x.head())
display(train_y.head())
```

```python
from xgboost import XGBClassifier
from xgboost import plot_importance
from matplotlib import pyplot
from numpy import sort
from sklearn.metrics import accuracy_score
from sklearn.feature_selection import SelectFromModel
```

```python
import warnings
warnings.filterwarnings("ignore")

model = XGBClassifier()
model.fit(train_x, train_y)
predictions = model.predict(test_x)
accuracy = accuracy_score(test_y, predictions)
print("Accuracy: %.2f%%" % (accuracy * 100.0))

# Fit model using each importance as a threshold
thresholds = sort(model.feature_importances_)
for thresh in thresholds:
    selection = SelectFromModel(model, threshold=thresh, prefit=True)
    select_X_train = selection.transform(train_x)
    # train model
    selection_model = XGBClassifier()
    selection_model.fit(select_X_train, train_y)
    # eval model
    select_X_test = selection.transform(test_x)
    predictions = selection_model.predict(select_X_test)
    accuracy = accuracy_score(test_y, predictions)
    print("Thresh=%.3f, n=%d, Accuracy: %.2f%%" % (thresh, select_X_train.shape[1], accuracy*100.0))




plot_importance(model)
pyplot.show()
```
